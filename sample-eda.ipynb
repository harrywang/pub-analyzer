```python
import sqlite3
import pandas as pd
```

```python
# direct connect to sqlite3
conn = sqlite3.connect("data/pub_raw.db")
cur = conn.cursor()
cur.execute("select * from author limit 5;")
results = cur.fetchall()
print(results)
cur.close()
conn.close()
```

```python
# connect and return pandas dataframe
conn = sqlite3.connect("data/pub_raw.db")
authors = pd.read_sql_query("select * from author limit 5;", conn)
authors
```

```python
papers = pd.read_sql_query("select * from paper limit 5;", conn)
papers
```
